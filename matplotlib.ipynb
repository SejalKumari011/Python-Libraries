# Title

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

Changing color

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='Black')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

Changing color using color code

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

## Changing Line styles
