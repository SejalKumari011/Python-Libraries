## Changing Line styles
### 1. Dashed
### 2. Dotted

### 1. Dashed

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dashed')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

### 2. Dotted

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dotted')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

### 3. dashdot

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dashdot')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

### Line width

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='solid',linewidth=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='solid',linewidth=3)
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

### Highlight the Points using Marker

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dashdot',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dashdot',marker='D')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

Legend

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dashdot',marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dashdot',marker='D',label='Rohit')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='best')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#14142d',linestyle='dashdot',marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#a728c0',linestyle='dashdot',marker='D',label='Rohit')
plt.title("Virat Kohli Vs. Rohit Sharma Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='best')

plt.grid()
plt.show()
```

```python
plt.barh(color,children,color='black')
```

```python
df=pd.read_csv("batsman_season_record.csv")
df
x=np.arange(df.shape[0])
plt.bar(x-0.2,df['2015'],width=0.2)
plt.bar(x,df['2016'],width=0.2)
plt.bar(x+0.2,df['2016'],width=0.2)
plt.xticks(x,df['batsman'])
plt.show()

```

```python
color=['Red  red red red red','yellow,yellow, yellow, yellow','green green green','pink pink pink']
children=[10,5,8,15]
plt.bar(color,children,color='orange')
plt.xticks(rotation='vertical')
```

### Stacked Bar Chart

```python
df=pd.read_csv("batsman_season_record.csv")
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
plt.legend()
plt.show()
```
