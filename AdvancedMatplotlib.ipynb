```python
fig,ax=plt.subplots(2,1,sharex=True,figsize=(10,6))
ax[0].scatter(df['avg'],df['runs'], color='red')
ax[1].scatter(df['avg'],df['strike_rate'])
ax[0].set_title('Runs vs Avg Graph')
ax[0].set_ylabel('Runs')
ax[1].set_title('Avg vs Strike rate')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')

```

## 3D Plots

### 3D Line plot

```python
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]

fig=plt.figure()

ax=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100])  #s=marker sixe
ax.plot3D(x,y,z,color='red')    #3d line
```

### 3D surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
xx.shape
```

```python
z=xx**2+yy**2    #XX^2+YY^2
```

```python
fig=plt.figure(figsize=(10,6))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.log(yy)
fig=plt.figure(figsize=(10,6))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Converting 3d graph to 2d using contour 

```python

fig=plt.figure(figsize=(10,6))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
