```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv('iris.csv')
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa': 0, 'Iris-virginica': 1, 'Iris-versicolor': 2})
```

```python
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Lengeth')
plt.colorbar()
```

### size of the graph

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Lengeth')
plt.colorbar()
```

### Anotation

```python
df=pd.read_csv('batter.csv')
df.head()
```

```python
sample_df=df.head(100).sample(25,random_state=5)
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
x=[1,2,3,4]
y=[5,7,9,4]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
```

### Giving text to each point on the graph

```python
sample_df.head()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
    
```
