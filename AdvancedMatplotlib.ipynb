### Plot multiple Graphs together side by side

```python
fig,ax=plt.subplots(2,1,sharex=True,figsize=(10,6))
ax[0].scatter(df['avg'],df['runs'], color='red')
ax[1].scatter(df['avg'],df['strike_rate'])
ax[0].set_title('Runs vs Avg Graph')
ax[0].set_ylabel('Runs')
ax[1].set_title('Avg vs Strike rate')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')

```

```python
fig,ax=plt.subplots(2,2,figsize=(10,6))

ax[0,0].scatter(df['avg'],df['runs'], color='red')

ax[0,1].scatter(df['avg'],df['strike_rate'])

# ax[0,0].set_title('Runs vs Avg Graph')
# ax[0,0].set_ylabel('Runs')

# ax[0,1].set_title('Avg vs Strike rate')
# ax[0,1].set_xlabel('Average')
# ax[0,1].set_ylabel('Strike Rate')

ax[1,0].hist(df['avg'],color='red')

ax[1,1].hist(df['runs'])

# ax[1,0].set_title('Avg ')

# ax[1,1].set_title('Runs ')
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['runs'], color='red')

ax2=fig.add_subplot(2,2,2)
ax2.hist(df['avg'],color='red')

ax3=fig.add_subplot(2,2,3)
ax3.hist(df['runs'])

```
