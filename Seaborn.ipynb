```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

Loading tips dataset from seaborn

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

### scatter plot --> axis plot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex')
```

we can make axis plotto figure plot using relplot

```python
sns.relplot(data=tips, x="total_bill", y='tip', kind='scatter',hue='sex',style='time',size='size')
```

loading library from plotly

```python
gap=px.data.gapminder()
gap.head()
```

```python
temp_df=gap[gap['country']=='India']
```

### Line plot

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

Using relplot

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
#hue
temp_df=gap[gap['country'].isin(['India','America','Brazil'])]
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

facet plot col basis

```python
sns.relplot(data=tips, x="total_bill",y="tip",kind="scatter",col='sex',row='day')
```

```python
sns.relplot(data=gap,x="lifeExp",y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```
